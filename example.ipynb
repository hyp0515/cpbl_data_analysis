```python
import json
import matplotlib.pyplot as plt
```

```python
# 讀取 2023 上半季資料成 rebas_open_data (存放資料的資料夾須跟程式檔案在同一個資料夾裡)
with open('/Users/hyp0515/data/CPBL-2024-OpenData/CPBL-2024-OpenData.json') as f:
    rebas_open_data = json.load(f)
```

```python
coordx_list = []
coordy_list = []
velo_list = []



for i in range(len(rebas_open_data[0]['homePAList'][0]['events'])):
    coordx, coordy = (rebas_open_data[0]['homePAList'][0]['events'][i]['coordX']), (rebas_open_data[0]['homePAList'][0]['events'][i]['coordY'])
    print(f"coordX: {coordx}, coordY: {coordy}")
    coordx_list.append(float(coordx))
    coordy_list.append(float(coordy))
    pitchtype = rebas_open_data[0]['homePAList'][0]['events'][i]['pitchType']
    print(f"pitchType: {pitchtype}")
    velo = rebas_open_data[0]['homePAList'][0]['events'][i]['velocity']
    print(f"velocity: {velo}")
    velo_list.append(float(velo))
    print("-----")
```
